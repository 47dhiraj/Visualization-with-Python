```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataframe = pd.DataFrame({
    "Languages": ["Python", "JavaScript", "Java", "Go", "C++"],
    "Popularity": [85, 70, 60, 55, 50]
})
```

```python
sns.barplot(data=dataframe, 
            x='Popularity',  
            y='Languages',   
            palette='husl',
            hue='Languages',  
            # orient='h',      
            width=0.6,
           )

plt.xlabel("Popularity", fontsize=11)
plt.ylabel("Programming Languages", fontsize=11)
plt.title("Top 5 Popular Languages", fontsize=14)

# Show the plot
plt.show()
```
