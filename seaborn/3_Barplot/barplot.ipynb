### `Barplot`

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataframe = sns.load_dataset('healthexp')
dataframe
```

```python
## Accessing latest Year data

## using ==>pandas query
# latest_year = dataframe.query('Year == 2020')

## using ==> Boolean indexing
latest_year = dataframe[dataframe['Year'] == 2020]
latest_year
```

```python
sns.barplot(data=latest_year, 
            x='Country', 
            y='Spending_USD', 
            hue='Country', 
            saturation=0.9, 
            alpha=0.9)


plt.show()
```
