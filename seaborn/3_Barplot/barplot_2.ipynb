```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataframe = pd.DataFrame({
    "Languages": ["Python", "JavaScript", "Java", "Go", "C++"],
    "Popularity": [85, 70, 60, 55, 50]
})
```

```python
plt.figure(figsize=(8, 6))

sns.barplot(data=dataframe, 
            x='Languages', 
            y='Popularity', 
            palette='husl',
            hue='Languages'
           )

plt.xlabel("Programming Languages", fontsize=11)
plt.ylabel("Popularity", fontsize=11)
plt.title("Top 5 Popular Languages", fontsize=14)

plt.show()
```
