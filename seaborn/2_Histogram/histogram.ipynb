### `Histogram`

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset("flights")

df
```

```python
# Grouping data bye year and calculate total passengers
if df['year'].dtype.name == 'category':
    df['year'] = df['year'].astype(int)
```

```python
yearly_data = df.groupby('year', as_index=False)['passengers'].sum()
```

```python
yearly_data
```

```python
sns.histplot(
    data=yearly_data,
    x='year',
    weights='passengers',
    # bins=bin_edges,
    bins='auto',
    discrete=True,
    kde=True,
    color='green',
)

plt.gca().lines[0].set_color('darkblue')
plt.gca().lines[0].set_linewidth(3)       
plt.gca().lines[0].set_linestyle('--')    

plt.title('Total Passengers per Year', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Total Passengers', fontsize=12)

plt.show()
```
