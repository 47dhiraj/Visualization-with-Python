```python
import numpy as np
import matplotlib.pyplot as plt

frequency = 440         
amplitude = 1.0          
duration = 0.01         
sampling_rate = 44100   

t = np.linspace(0, duration, int(sampling_rate * duration), endpoint=False)

# Formula: y(t)=A⋅sin(2πft) ==> to generate the sound wave data (sine wave)
sound_wave = amplitude * np.sin(2 * np.pi * frequency * t)

plt.style.use('dark_background')

plt.figure(figsize=(10, 4))
plt.plot(t, sound_wave, label=f"{frequency} Hz Sound Wave")

plt.title("Sound Wave")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")

plt.legend(loc=1)

plt.grid(False)

plt.tight_layout()
plt.show()
```
