### `Line Plot`

```python
import matplotlib.pyplot as plt
```

```python
print(plt.style.available)
```

```python
# EXAMPLE 1: sample temperature data of a week

dates = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']

temperatures = [25, 28, 27, 30, 29, 32, 31]

plt.figure(figsize=(8, 5))
plt.plot(dates, temperatures, label='Daily Temperature', color='blue', marker='o', linestyle='-', linewidth=2)

plt.title('Weekly Temperature Trend')
plt.xlabel('Day of the Week')
plt.ylabel('Temperature (C)')

plt.legend()

plt.grid(True, linestyle='--', alpha=1)

plt.tight_layout()
plt.show()
```

```python
# Example 2: Sample data of monthly sales figures (in thousands of Rupees)

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

sales = [16.5, 18.3, 17.8, 63.6, 70.4, 65.7, 62.3, 56.1, 42.5, 31.2, 29.7, 22.5]

plt.style.use('dark_background')
plt.figure(figsize=(9, 5))
plt.plot(months, sales, label='Monthly Sales', color='orange', marker='o', linestyle='-')

plt.xlabel('Month')
plt.ylabel('Sales (in Rs.1000)')
plt.title('Monthly Sales Trend in 2024')
plt.grid(True, linestyle='-', alpha=0.7)

plt.legend()

plt.show()
```
