### Stack plot aka `Area Plot`

```python
import matplotlib.pyplot as plt
```

```python
plt.style.use("ggplot")
```

```python
Days = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]

freelancer = [7, 6, 2, 6, 3, 6, 7]
software_developer = [8, 8, 8, 8, 8, 0, 0]
chef = [0, 0, 8, 8, 8, 10, 10]

labels = ['Software Developer', 'Chef', 'Freelancer']
colors = ['#6d904f', '#fc4f30', '#008fd5']


plt.stackplot(Days, software_developer, chef, freelancer, labels=labels, colors=colors)

plt.legend(loc=1)

plt.xlabel("Days of Week", fontsize="13")
plt.ylabel("Working Hours", fontsize="13")
plt.title("Professional Working Hours") 

plt.tight_layout()
plt.show()
```
