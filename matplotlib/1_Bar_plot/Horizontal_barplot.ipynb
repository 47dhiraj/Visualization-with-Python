### `Horizontal Barplot`

```python
import matplotlib.pyplot as plt
```

```python
print(plt.style.available)
```

```python
plt.style.use('ggplot')
```

```python
x = ["Python", "JavaScript", "Java", "Go", "C++"]

y = [85, 70, 60, 55, 50]

plt.xlabel("Programming Languages", fontsize="13")
plt.ylabel("Popularity", fontsize="13")
plt.title("Top 5 Popular Languages")

c = ["blue", "yellow", "orange", "magenta", "green"]

for i in range(len(x)):
    plt.barh(x[i], y[i], height=0.5, color=c[i], label=x[i], align="edge", alpha=0.8)

plt.legend(title="Languages")

plt.grid(True)
plt.tight_layout()
plt.show()
```

### `For showing bars in reverse order`

```python
plt.xlabel("Programming Languages", fontsize="13")
plt.ylabel("Popularity", fontsize="13")
plt.title("Top 5 Popular Languages")

c = ["blue", "yellow", "orange", "magenta", "green"]

for i in range(len(x)):
    plt.barh(x[i], y[i], height=0.5, color=c[i], label=x[i], align="edge", alpha=0.8)
    
plt.gca().invert_yaxis()


plt.legend(title="Languages")

plt.grid(False)
plt.tight_layout()
plt.show()
```
