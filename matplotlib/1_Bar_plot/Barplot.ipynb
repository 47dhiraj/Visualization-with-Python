## `Bar plot`

#### To get detail info of any function ` ==> shift + Tab `

```python
import matplotlib.pyplot as plt
```

```python
print(plt.style.available)
```

```python
plt.style.use('fivethirtyeight')
```

```python
x = ["Python", "JavaScript", "Java", "Go", "C++"]
y = [85, 70, 60, 55, 50]
```

#### `simple Bar Plot`

```python
plt.bar(x, y)
plt.show()
```

```python
plt.xlabel("Programming Languages", fontsize="13")
plt.ylabel("Popularity", fontsize="13")

plt.title("Top 5 Popular Languages")


plt.bar(x, y, width=0.5)

plt.show()
```

```python
plt.xlabel("Programming Languages", fontsize="11")
plt.ylabel("Popularity", fontsize="11")

plt.title("Top 5 Popular Languages", fontsize="12")

c = ["blue", "yellow", "orange", "magenta", "green"]

plt.bar(x, y, width=0.5, color=c, edgecolor="grey", linewidth=3)     
plt.show()
```

```python
plt.xlabel("Programming Languages", fontsize="11")
plt.ylabel("Popularity", fontsize="11")

plt.title("Top 5 Popular Languages", fontsize="12")

c = ["blue", "yellow", "orange", "magenta", "green"]

plt.bar(x, y, width=0.5, color=c, align="edge")
plt.show()
```

```python
plt.xlabel("Programming Languages", fontsize="11")
plt.ylabel("Popularity", fontsize="11")

plt.title("Top 5 Popular Languages", fontsize="12")

c = ["blue", "yellow", "orange", "magenta", "green"]

plt.bar(x, y, width=0.5, color=c, align="edge", alpha=0.8)

plt.tight_layout()
plt.show()
```

### `Adding labels for legend`

```python
plt.xlabel("Programming Languages", fontsize="11")
plt.ylabel("Popularity", fontsize="11")
plt.title("Top 5 Popular Languages", fontsize="12")

c = ["blue", "yellow", "orange", "magenta", "green"]


for i in range(len(x)):
    plt.bar(x[i], y[i], color=c[i], label=x[i], width=0.5, align="edge", alpha=0.8)

plt.legend(title="Languages", fontsize=10)

plt.tight_layout()
plt.grid(False)
plt.show()
```
