### `Grouped / side-by-side Bar Plot`

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
print(plt.style.available)
```

```python
plt.style.use('ggplot')
```

```python
x = ["Python", "JavaScript", "Java", "Go", "C++"]

y1 = [45, 60, 85, 40, 75]
y2 = [90, 75, 40, 60, 30]

x_index = np.arange(len(x))

width = 0.3

plt.bar(x_index - width/2, y1, width, label='Popularity 1')
plt.bar(x_index + width/2, y2, width, label='Popularity 2')

plt.xlabel("Programming Languages", fontsize="13")
plt.ylabel("Popularity", fontsize="13")
plt.title("Top 5 Popular Languages")

plt.legend()

plt.xticks(x_index, x, rotation=10)

plt.tight_layout()
plt.show()
```
