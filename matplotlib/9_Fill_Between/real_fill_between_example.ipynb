```python
import pandas as pd

from matplotlib import pyplot as plt
# import matplotlib.pyplot as plt
```

```python
print(plt.style.available)
```

```python
plt.style.use("ggplot")
# plt.style.use("seaborn-v0_8-notebook")
```

```python
# reading data from csv & storing to dataframe
data = pd.read_csv('data.csv')

## getting each column from dataframe
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']

plt.figure(figsize=(7.5, 5))

# creates a plot
plt.plot(ages, dev_salaries, color='#444444',
         linestyle='--', label='All Developer')

# creates a plot
plt.plot(ages, py_salaries, label='Python Developer')


plt.fill_between(ages, py_salaries, dev_salaries,
                 where=(py_salaries > dev_salaries),
                 interpolate=True,
                 color='green',
                 alpha=0.25, 
                 label='Salary above average')

plt.fill_between(ages, py_salaries, dev_salaries,
                 where=(py_salaries <= dev_salaries),
                 interpolate=True,
                 color='red', 
                 alpha=0.25, 
                 label='Salary below Avg')

plt.legend()

plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')

plt.tight_layout()

plt.show()
```
