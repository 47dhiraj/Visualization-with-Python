### `Subplot`

```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
plt.style.use("ggplot")
```

```python
x = [1, 2, 3, 4, 5]
y = [1, 2, 3, 4, 5]


plt.subplot(2, 2, 1)
plt.plot(x, y, color="blue")


plt.subplot(2, 2, 2)
plt.pie([100], colors="red", wedgeprops={'edgecolor': 'black', 'linewidth': 1, 'width': 0.5})


plt.subplot(2, 2, 3)
pie_values = [10, 20, 30, 40]
plt.pie(pie_values, radius=1.2, wedgeprops={'edgecolor': 'black', 'linewidth': 1})


x1 = ["C#", "Java", "JavaScript", "Python"]
y1 = [40, 60, 75, 90]
plt.subplot(2, 2, 4)
plt.bar(x1, y1)


plt.tight_layout()
plt.show()
```
