### `Subplots`

```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
plt.style.use('ggplot')

data = pd.read_csv('data.csv')
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']

fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1, figsize=(10, 8))

ax1.plot(ages, dev_salaries, color='#444444', linestyle='--', label='All Devs')
ax1.set_title('Developer Salaries by Age')
ax1.set_ylabel('Salary in USD')
ax1.legend()

ax2.plot(ages, py_salaries, label='Python', color='blue')
ax2.plot(ages, js_salaries, label='JavaScript', color='green')
ax2.set_title('Salary Comparison: Python vs JavaScript')
ax2.set_xlabel('Age')
ax2.set_ylabel('Salary in USD')
ax2.legend()

plt.tight_layout()

plt.show()
fig.savefig('salary.png')
```
