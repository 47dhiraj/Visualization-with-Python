## Boxplot ` aka  box-and-whisker plot`

```python
import matplotlib.pyplot as plt
```

```python
plt.style.use("ggplot")
```

### `Simple Boxplot`

```python
x = [10, 20, 30, 40, 50, 60, 70, 80, 90]

plt.boxplot(x)
plt.show()
```

### Finding the `Outliers`

```python
x = [10, 20, 30, 40, 50, 60, 70, 80, 90, 200]

plt.boxplot(x)
plt.show()
```
