### `Multiple Boxplots`

```python
import matplotlib.pyplot as plt
```

```python
plt.style.use('ggplot')
```

```python
# Munal section computer scores: 30 students
munal_scores = [80, 82, 85, 87, 89, 90, 91, 92, 94, 96, 97, 98, 99, 95, 88, 93, 91, 87, 84, 92, 86, 91, 90, 89, 92, 85, 93, 94, 88, 85]  # Munal's 30 students

# Munal section computer scores: 30 students
danfe_scores = [99, 85, 75, 82, 70, 90, 91, 92, 94, 96, 87, 98, 99, 95, 88, 83, 91, 87, 74, 92, 86, 91, 90, 89, 60, 70, 72, 45, 75, 63]  # Danfe's 30 students

data = [munal_scores, danfe_scores]

plt.figure(figsize=(7, 5))

plt.boxplot(data, 
            patch_artist=True,
            showmeans=True,
            boxprops=dict(facecolor='lightblue', color='black'), 
            medianprops=dict(color='red', linewidth=2))

# Customize the plot (optional)
plt.title('SEE Computer Science score comparison: Munal vs Danfe', fontsize=10)

# Naming x-axis labels
plt.xticks([1, 2], ['Munal', 'Danfe'])  

# Show the plot
plt.tight_layout()
plt.show()
```
