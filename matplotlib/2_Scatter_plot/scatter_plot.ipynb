### `Scatter Plot`

#### For knowing details of function ==> `Shift + Tab`

```python
import matplotlib.pyplot as plt
```

```python
plt.style.use('fivethirtyeight')
```

#### `Simple Scatter Plot`

```python
days = [1, 2, 3, 4, 5, 6, 7]
hours = [2, 4, 7, 9, 6, 4, 1]

plt.xlabel("Days", fontsize=12)
plt.ylabel("Hours", fontsize=12)
plt.title("Daily Working Hours - Scatter Plot", fontsize=12)

plt.scatter(days, hours)
plt.show()
```

#### Giving `color, sizes` to the scatter dots

```python
color = ["red", "orange", "green", "blue", "magenta", "yellow", "purple"]
sizes = [100, 200, 300, 500, 400, 200, 100]

plt.scatter(days, hours, c=color, s=sizes)
plt.show()
```

### Adding transparency using `alpha`

```python
plt.scatter(days, hours, c=color, s=sizes, alpha=0.7)
plt.show()
```

### using different `marker`

```python
plt.scatter(days, hours, c=color, s=sizes, alpha=0.7, marker="o")
plt.show()
```

```python
plt.scatter(days, hours, c=color, s=sizes, alpha=0.7, marker="*")
plt.show()
```

```python
plt.scatter(days, hours, c=color, s=sizes, alpha=0.7, marker="s")
plt.show()
```

```python
plt.scatter(days, hours, c=color, s=sizes, alpha=0.7, marker="^")
plt.show()
```

```python
plt.scatter(days, hours, c=color, s=sizes, alpha=0.7, marker="+")
plt.show()
```

### Use of `cmap  i.e color map`

```python
import numpy as np

color_values = np.random.randint(1, 900, size=7)

plt.scatter(days, hours, c=color_values, s=sizes, alpha=0.8, cmap="coolwarm")

plt.show()
```

### Adding `colorbar`

```python
color_values = np.random.randint(1, 900, size=7)

plt.scatter(days, hours, c=color_values, s=sizes, alpha=0.8, cmap="coolwarm")
plt.colorbar(label='Color Intensity')

plt.show()
```
